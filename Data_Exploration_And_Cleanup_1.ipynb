```python
# Check whether we have the same number of data
df3.count()
```

```python
# Clean Data
df4 = df3.dropna(how='any')
```

```python
df4.count()
```

```python
# Append year and month into new columns
new_year = []
new_month = []
for row in df4['Fiscal Year']:
    y = str(row)[:4]
    m = str(row)[4:6]
    new_year.append(y)
    new_month.append(m)
#print(new_year)
#print(new_month)

se = pd.Series(new_year)
see = pd.Series(new_month)

df4['New Year'] = se.values
df4['New Month'] = see.values

df4.head()
```

```python
# Insert into a new DataFrame
new_pd = pd.DataFrame(df4, columns=[
                        "Fiscal Year", "Revenue", "New Year", "New Month"])
new_pd
```

```python
# Define the day base on the month since not available 
new_day = []
for row in new_pd['New Month']:
    if row == '12':
        s = '31'
        new_day.append(s)
    elif row == '09':
        s = '30'
        new_day.append(s)
    elif row == '06':
        s = '30'
        new_day.append(s)
    elif row == '03':
        s = '31'
        new_day.append(s)
#print(new_day)
sed = pd.Series(new_day)        
```

```python
# Add New Day into DataFrame
df4['New Day'] = sed.values

df4
```

```python
# Reorganize DataFrame
new_pd = pd.DataFrame(df4, columns=[
                        "Fiscal Year", "Revenue", "New Year", "New Month", "New Day"])
new_pd.head()
```

```python
# Check type of data
new_pd.dtypes
```

```python
# Convert data into numeric
new_pd['New Year'] = pd.to_numeric(new_pd['New Year'], errors='coerce')
new_pd['New Month'] = pd.to_numeric(new_pd['New Month'], errors='coerce')
new_pd['New Day'] = pd.to_numeric(new_pd['New Day'], errors='coerce')
new_pd.head()
```

```python
# Check type of data
new_pd.dtypes
```

```python
# Create a new date based on three columns
final_date = []
s = "-"
index = new_pd.index.values

for x in index:
    n = str(new_pd['New Year'].loc[x]) + s + str(new_pd['New Month'].loc[x]) + s + str(new_pd['New Day'].loc[x])
    final_date.append(n)
    #print(n)
#print(final_date)
sd = pd.Series(final_date)

new_pd['New Date'] = sd.values
new_pd.head()
```

```python
# Reorganize DataFrame
new_pd_final = pd.DataFrame(new_pd, columns=[
                        "New Date", "Revenue"])
new_pd_final.head()
```

```python
# Sort data
test = new_pd_final.sort_values(by='New Date')
test
index1 = test.index.values
```
