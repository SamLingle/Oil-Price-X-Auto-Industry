```python
new_year = []
new_month = []
for row in df4['Fiscal Year']:
    y = str(row)[:4]
    m = str(row)[4:6]
    new_year.append(y)
    new_month.append(m)
#print(new_year)
#print(new_month)

se = pd.Series(new_year)
see = pd.Series(new_month)

df4['New Year'] = se.values
df4['New Month'] = see.values

df4.head()
```

```python
new_pd = pd.DataFrame(df4, columns=[
                        "Fiscal Year", "Revenue", "New Year", "New Month"])
new_pd
```

```python
new_day = []
for row in new_pd['New Month']:
    if row == '12':
        s = '31'
        new_day.append(s)
    elif row == '09':
        s = '30'
        new_day.append(s)
    elif row == '06':
        s = '30'
        new_day.append(s)
    elif row == '03':
        s = '31'
        new_day.append(s)
#print(new_day)
sed = pd.Series(new_day)        
```
