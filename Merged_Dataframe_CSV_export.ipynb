```python
# Dependencies
import pandas as pd
from functools import reduce
```

```python
# Store filepaths in variables.
honda_csv = "Honda.csv"
ford_csv = "Ford.csv"
gm_csv = "GM.csv"
toyota_csv = "Toyota.csv"
oil_csv = "WTI_Crude_Oil.csv"
```

```python
honda = pd.read_csv(honda_csv)
honda = honda.rename(columns={'Last Price':'Honda stock price'})
honda.head()
```

```python
gm = pd.read_csv(gm_csv)
gm = gm.rename(columns={'Last Price':'GM stock price'})
gm.head()
```

```python
toyota = pd.read_csv(toyota_csv)
toyota = toyota.rename(columns={'Last Price':'Toyota stock price'})
toyota.head()
```

```python
ford = pd.read_csv(ford_csv)
ford = ford.rename(columns={'Last Price':'Ford stock price'})
ford.head()
```

```python
WTI_crude_oil = pd.read_csv(oil_csv)
WTI_crude_oil = WTI_crude_oil.rename(columns={'Last Price':'Crude Oil stock price'})
WTI_crude_oil.head()
```

```python
# compile the list of dataframes want to merge
data_frames = [WTI_crude_oil, ford, honda, gm, toyota]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Dates'],
                                            how='inner'), data_frames).fillna('void')
df_merged.head()
```

```python
# Reorganize and rename the data frame.
df_merged_organized = df_merged[['Dates', 'Crude Oil stock price','Ford stock price','Honda stock price','Toyota stock price','GM stock price']]
df_merged_organized.head()
```

```python
pd.DataFrame.to_csv(df_merged_organized, 'merged_stock_prices.txt', sep=',', na_rep='.', index=False)
```
